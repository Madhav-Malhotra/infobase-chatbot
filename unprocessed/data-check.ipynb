```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('./articles.json')
df.info()
```

```python
def content_len_bar_graph():
    # Compute content lengths
    df['content_lengths'] = df['content'].apply(len)
    condensed_links = df['link'].map(lambda x : x.split('/')[-1].split('.')[0])

    # Setup plot axes (vert axis has links, horiz axis has lengths)
    fig, ax = plt.subplots()
    ax.barh(condensed_links, df['content_lengths'], label=list(df['content_lengths']))
    ax.set_xlabel('Content Length')
    ax.set_ylabel('Article Title')
    ax.set_title('Article Content Lengths')

    # Increase plot height
    fig.set_size_inches(10, 20)

    # Add labels
    ax.bar_label(ax.containers[0], label_type='edge')

    # Show plot
    plt.show()

content_len_bar_graph()
    
```
